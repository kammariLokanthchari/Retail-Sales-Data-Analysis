# <b> Retail Sales Data Analysis Using Python <b>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("Sample_ Superstore.csv")

df.head()
```

```python
print("Total Sales =", df["Sales"].sum())
```

```python
top_products = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False)

top_products.head(10)
```

```python
top_products.head(10).plot(kind="bar")

plt.title("Top 10 Products by Sales")
plt.ylabel("Sales")
plt.show()
```

```python
df.isnull().sum()
```

```python
print("Total Sales:", df["Sales"].sum())
print("Total Profit:", df["Profit"].sum())
```

```python
category_sales = df.groupby("Category")["Sales"].sum()

print(category_sales)
```

```python
category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.ylabel("Sales")
plt.show()
```

```python
category_profit = df.groupby("Category")["Profit"].sum()

category_profit.plot(kind="bar")

plt.title("Profit by Category")
plt.ylabel("Profit")
plt.show()
```

```python
top_states = df.groupby("State")["Sales"].sum().sort_values(ascending=False)

top_states.head(10)
```

```python
top_profit_products = df.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False)

print(top_profit_products.head(10))
```

```python
import pandas as pd

df["Order Date"] = pd.to_datetime(
    df["Order Date"],
    format="mixed",
    dayfirst=True
)
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])

df["Month"] = df["Order Date"].dt.to_period("M")

monthly_sales = df.groupby("Month")["Sales"].sum()

monthly_sales.plot(figsize=(12,5))

plt.title("Monthly Sales Trend")
plt.ylabel("Sales")
plt.show()
```
